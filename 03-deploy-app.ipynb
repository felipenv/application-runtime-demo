## Deploy vizro dashboard

```python
import mlrun
```

```python
project = mlrun.get_or_create_project("application-runtime", "./", user_project=True, )
```

### Source files from git

```python
project.set_source(source="git://github.com/felipenv/application-runtime-demo.git#main", pull_at_runtime=False)
```

### set function with kind application

```python
# Specify source to be loaded on build time or run time
application = project.set_function(name="vizro-map", kind="application", image="mlrun/mlrun", requirements_file="./requirements.txt", with_repo=True,)
project.save()
```

### define command for container

```python
application.set_internal_application_port(8050)
application.spec.command = "gunicorn"
application.spec.args = [
    "--chdir",
    "/home/mlrun_code",
    "viz_app:app",
    "--bind",
    "0.0.0.0:8050",
    "--log-level",
    "debug",
]
```

### Deploy app

```python
application.deploy(with_mlrun=False)
```

```python
application.invoke("/", verify=False)
```
